```python
from openai import OpenAI
client = OpenAI(api_key="")
```

```python
# 원본 (임금님) 데이터셋 20개
light_fine_tuned_model_id = "ft:gpt-3.5-turbo-1106:choi::"
fine_tuned_model_id = "ft:gpt-3.5-turbo-1106:choi::"
model_id = "gpt-3.5-turbo-1106"

completion = client.chat.completions.create(
    model=light_fine_tuned_model_id,  # 20개 모델
    messages=[
        {"role": "system", "content": "임금의 말투를 사용하여 답장하시오."},
        {"role": "user", "content": "오늘 점심은 뭐 먹을래?"}
    ]
)

print(completion.choices[0].message.content)
```

```python
# 원본 (임금님) 데이터셋 100개

# Use a general model for chat completion
completion = client.chat.completions.create(
    model=fine_tuned_model_id,  # 100개 모델
    messages=[
        {"role": "system", "content": "임금의 말투를 사용하여 답장하시오."},
        {"role": "user", "content": "오늘 점심은 뭐 먹을래?"}
    ]
)

print(completion.choices[0].message.content)
```

```python
# 변경 (아이유) 데이터셋 20개
light_fine_tuned_model_id_iu = "ft:gpt-3.5-turbo-1106:choi::"
fine_tuned_model_id_iu = "ft:gpt-3.5-turbo-1106:choi::"
model_id = "gpt-3.5-turbo-1106"

system_prompt = "아이유처럼 귀엽고 부드럽고 예쁘고 착한 말투로 답장하시오."

# 10개 이상의 창의적인 질문 리스트
creative_questions = [
    "오늘 기분은 어때?",
    "만약 오늘을 향기로 표현하면 어떤 향일까?",
    "지금 듣고 싶은 노래 있어?",
    "오늘 하루에서 가장 고마웠던 순간은 뭐야?",
    "하늘에 별 하나 가져올 수 있다면 어떤 별일까?",
    "오늘 나를 따뜻하게 했던 일이 있었어?",
    "만약 지금 바로 여행 갈 수 있다면 어디로 가고 싶어?",
    "지금 마음속에서 가장 반짝이는 생각은 뭐야?",
    "오늘 스스로에게 해주고 싶은 말은 있어?",
    "지금 창밖을 보면 어떤 풍경이 있으면 좋겠어?",
    "오늘 밤에 소원을 하나 빌 수 있다면 뭐라고 빌래?"
]

print("20개 데이터셋 아이유 모델 테스트")

for idx, question in enumerate(creative_questions, start=1):
    completion = client.chat.completions.create(
        model=light_fine_tuned_model_id_iu,   # 20개 모델
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": question}
        ]
    )

    print(f"[20개 - {idx}] User: {question}")
    print(f"[20개 - {idx}] IU: {completion.choices[0].message.content}")
    print("-" * 60)

```

```python
# 변경 (아이유) 데이터셋 100개
print("100개 데이터셋 아이유 모델 테스트")

for idx, question in enumerate(creative_questions, start=1):
    completion = client.chat.completions.create(
        model=fine_tuned_model_id_iu,   # 100개 모델
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": question}
        ]
    )

    print(f"[100개 - {idx}] User: {question}")
    print(f"[100개 - {idx}] IU: {completion.choices[0].message.content}")
    print("-" * 60)
```
